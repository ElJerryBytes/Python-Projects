```python
import random

class JuegoAdivinanza:
    def __init__(self):
        self.palabras = ["python", "programacion", "objetos", "divertido", "aprendizaje", "metodos" , "atributos"]
        self.palabra_secreta = random.choice(self.palabras)
        self.intentos_maximos = 5
        self.intentos_restantes = self.intentos_maximos
        self.letras_adivinadas = ["_"] * len(self.palabra_secreta)

    def mostrar_estado(self):
        print("Palabra secreta:", " ".join(self.letras_adivinadas))
        print("Intentos restantes:", self.intentos_restantes)
        print()

    def adivinar_letra(self, letra):
        if letra in self.palabra_secreta:
            for i in range(len(self.palabra_secreta)):
                if self.palabra_secreta[i] == letra:
                    self.letras_adivinadas[i] = letra
        else:
            self.intentos_restantes -= 1

    def advinar_palabra(self, palabra):
        if palabra == self.palabra_secreta:
            self.letras_adivinadas = list(palabra)
        else:
            self.intentos_restantes -= 1

juego = JuegoAdivinanza()
while True:
    juego.mostrar_estado()
    letra = input("Ingresa una letra: ").lower()
    juego.adivinar_letra(letra)

    if "_" not in juego.letras_adivinadas:
        print("¡Felicidades! Adivinaste la palabra:", juego.palabra_secreta)
        print()
        juego.palabra_secreta = random.choice(juego.palabras) #El juego continua
        juego.intentos_restantes = juego.intentos_maximos
        juego.letras_adivinadas = ["_"] * len(juego.palabra_secreta)

    if juego.intentos_restantes == 0:
        print("Te quedaste sin intentos. Por favor vuelve a intentarlo. La palabra secreta era:", juego.palabra_secreta)
        break
```
